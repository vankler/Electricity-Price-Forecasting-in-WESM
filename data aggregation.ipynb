# Edit Data for Data Preparation

```python
# changing hours interval from 1-24 to 0-23
# to suit when parsing dates
hours = data['Interval']
new_hours = []
for hour in hours:
  new_hour = hour - 1
  new_hours.append(new_hour)
```

```python
hours = pd.Series(new_hours)
```

```python
data['Interval'] = hours
```

```python
data
```

```python
# removes the comma in values i.e. 3,123.092 to 3123.092
# commas make values object type instead of float
data = data.replace(',', '', regex=True)
```

```python
data.dtypes
```

The values of `Effective GWAP` and `Rolling Average` is still of type object. However, removing commas ensures that it would be the right data type when loaded in another notebook.

```python
data.to_csv(path + 'data_edited.csv', index=False)
```
